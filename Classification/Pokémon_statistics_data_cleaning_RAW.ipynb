```python
# 2.4 First conclusions:

# Since this dataset contains 800 rows we don't want to drop Pokémon. This could make it more difficult to draw conclusions
# from the dataset.

# Possible solutions:
# "type2": replace NaN with None
# "percentage_male": replace NaN with None
# "height_m" and "weight_kg": these ones could be imputed by calculate the median replace NaN with median
```

```python
# Sort imports later --> just for the first time the documentation is some sort of messed up.
import numpy as np
```

```python
# Replace NaN with None (check if NaN is the appropriate wording) --> see Doc in OneDrive why we use inplace = True
df["type2"].fillna("None", inplace = True)

# Replace NaN with None in "percentage_male"
df["percentage_male"].fillna("None", inplace = True)

# Introduce a new feature "genderless". 1 if the Pokémon is genderless and 0 if it's not.
# df["genderless"] = np.where(df["percentage_male"].isna(), 1, "None")

# Fill the missing values with exact mean (not rounded)
df["height_m"].fillna(df["height_m"].mean(), inplace = True)
df["weight_kg"].fillna(df["weight_kg"].mean(), inplace = True)
```

```python
# Check, if it worked out:
df.head().T
```

```python
# Overall, it looks good.
# Let's confirm this by looking at some Pokémon in detail (especially a legendary and one with an alternative form).

confirm_list = ["Mew", "Golem"]
df[df["name"].isin(confirm_list)].T
```

```python
# Now, columns are cleaned from missing values.
# However, there's something about "capture_rate". df.info() states that the values are of type object. But, looking in detail
# only integer values are to be identified.
# Let's examine this further.
```

```python
for i in df.capture_rate:
    print(i, end = ", ") # Explanation --> see Doc
```

```python
# It seems that one Pokémon in particular has two capture rates. This is an important finding.
# Assumption: we go with the lower capture rate.

df["capture_rate"].replace({"30 (Meteorite)255 (Core)" : "30"}, inplace = True)
```

```python
# Finally, convert "capture_rate":

df["capture_rate"] = df["capture_rate"].astype("int")
df["capture_rate"].dtype
```

```python
# Let's do some final steps towards the cleaning process.
# Remove "japanese_name" as it provides no added value for our model and "pokedex_number" as this could negatively influence
# our model we're about to develop.

df.drop(columns = ["japanese_name", "pokedex_number"], axis = 1, inplace = True) # Explanation --> see Doc
```

```python
# Fix a typo:

df.rename(columns = {"classfication" : "classification"}, inplace = True)
```

```python
# Set a Pokémon's name right at the beginning of the dataset:

df.insert(0, "name", df.pop("name"))
```
